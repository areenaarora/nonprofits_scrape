```python
import requests
import json
```

```python
all_nonprofits_endpoint = f"https://projects.propublica.org/nonprofits/api/v2/search.json?c_code%5Bid%5D=&city=&ntee%5Bid%5D&q_all=&q_any=&q_none=&q_phrase=&state%5Bid%5D=TN"
```

```python
response = requests.get(all_nonprofits_endpoint)
```

```python
pages_range = [str(x) for x in range(100)]
# all_nonprofits_json_data = response.json()
all_orgs_ein_list = []

for page_number in pages_range:
    print(f"processing page {page_number}")
    page_endpoint = f"{all_nonprofits_endpoint}&page={page_number}"
    page_response = requests.get(page_endpoint)
    page_nonprofits_data = page_response.json()

    for org_json in page_nonprofits_data["organizations"]:
        ein = org_json["ein"]
        all_orgs_ein_list.append(ein)
    print(f"done {page_number}")
```

```python
print(len(all_orgs_ein_list))
```

```python
org_list_of_info_dicts = []
for index, org_ein in enumerate(all_orgs_ein_list):
    org_endpoint = f"https://projects.propublica.org/nonprofits/api/v2/organizations/{org_ein}.json"
    org_data = requests.get(org_endpoint).json()

    org_dict = {}
    org_dict["name"] = org_data["organization"]["name"]
    org_dict["address"] = org_data["organization"]["address"]
    org_dict["city"] = org_data["organization"]["city"]
    org_dict["zipcode"] = org_data["organization"]["zipcode"]
    org_dict["classification_codes"] = org_data["organization"]["classification_codes"]
    org_dict["tax_period"] = org_data["organization"]["tax_period"]
    org_dict["asset_amount"] = org_data["organization"]["asset_amount"]
    org_dict["income_amount"] = org_data["organization"]["income_amount"]
    org_dict["revenue_amount"] = org_data["organization"]["revenue_amount"]
    org_dict["ntee_code"] = org_data["organization"]["ntee_code"]

    org_list_of_info_dicts.append(org_dict)
    print(f"#{index+1}, org ein: {org_ein}")
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(org_list_of_info_dicts)
```
